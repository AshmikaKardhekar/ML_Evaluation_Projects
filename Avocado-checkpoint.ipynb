# 1) Predicting AveragePrice

```python
#seperating the target variable
x_1=avoc_dat.drop('AveragePrice',axis=1)
y_1=avoc_dat['AveragePrice']
```

```python
x_1.shape,y_1.shape
```

```python
#checking the values of x
x_1.head()
```

Earlier in this notebook we LabelEncoded the variable 'region' as it is of dtype 'object' and we wanted to check it's correlation with other variables.

But now we need to apply one hot encoder to it as the feature 'region' is not ordinal and all the categories are independent of each other.

We'll apply one-hot-encoding, but first we'll have to convert the column 'region' as type 'object' so that ,pandas could recognise it for encoding. 

```python
#converting dtype of 'region' from int64 to object.
x_1['region']=x_1['region'].astype('object')
```

```python
#encode 'region' in x_1
x_1=pd.get_dummies(x_1)
x_1.head()
```

```python
#standard scale input variables
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
x_1=pd.DataFrame(scale.fit_transform(x_1),columns=x_1.columns)
x_1.head()
```

```python
#train-test-split
import sklearn
from sklearn.model_selection import train_test_split
x_1_train,x_1_test,y_1_train,y_1_test=train_test_split(x_1,y_1,test_size=0.3,random_state=4)
```

```python
#importing all the base regression algorithms
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
#importing all the metrics and tuning methods for regression analysis
from sklearn.model_selection import cross_val_score,GridSearchCV
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
#selecting the algorithm which works best 
models=[LinearRegression(),Lasso(),Ridge(),ElasticNet(),SVR(),KNeighborsRegressor(),DecisionTreeRegressor()]
for m in models:
    m.fit(x_1_train,y_1_train)
    print(m)
    print('SCORE :',m.score(x_1_train,y_1_train))
    pred=m.predict(x_1_test)
    print('r2_score : ',r2_score(y_1_test,pred))
    print('mean absolute error :',mean_absolute_error(y_1_test,pred))
    print('mean squared error :',mean_squared_error(y_1_test,pred))
    print('root mean squared error :',np.sqrt(mean_squared_error(y_1_test,pred)))
    print('\n')
```

SVR, DecisionTree and KNN show good scores.

DecisionTreeRegressor is showing the learning score of 100% (this algorithm may cause overfitting),And KNeighborsRegressor is showing a score of 92.17% .SVR is showing 86.46%.

DecisionTreeRegressor has r2_Score of 0.76 ,And KNeighborsRegressor has r2_Score of 0.85.This means KNeighbors is fitting better than DecisionTree. SVR is showing 0.81. SVR is fitting better than Decision Tree but not as good as KNeighbors.

DecisionTree also has a little greater values of mean_absolute_error,mean_squared_error and root_mean_squared_error as compared to KNeighbors and SVR. KNeighbors is showing slightly better metric values than SVR. 

So,finally we know that KNeighborsRegressor works best!

```python
#tuning the hyperparameters of KNN for better performance using GridSearchCV
params = [{'n_neighbors': [2,3,4,5,6], 'weights': ['uniform','distance']}]
KNN=KNeighborsRegressor()
GCV=GridSearchCV(estimator=KNN,param_grid=params,cv=10)
GCV.fit(x_1_train,y_1_train)
```

```python
#getting best parameters
print('best parameters are : ',GCV.best_params_)
```

```python
#final training and testing with best parameters
KNN_best=KNeighborsRegressor(n_neighbors=3,weights='distance')
KNN_best.fit(x_1_train,y_1_train)
preds=KNN_best.predict(x_1_test)
print('SCORE :',KNN_best.score(x_1_train,y_1_train))
print('r2_score : ',r2_score(y_1_test,preds))
print('mean absolute error :',mean_absolute_error(y_1_test,preds))
print('mean squared error :',mean_squared_error(y_1_test,preds))
print('root mean squared error :',np.sqrt(mean_squared_error(y_1_test,preds)))
```

We got a tuned KNN algroithm now.The metrics returned improved values now!

```python
#check whether model is overfitting or not
scores=cross_val_score(KNN_best,x_1_train,y_1_train,cv=10)
print('mean score : ',scores.mean())
```

Therefore we can be sure that the model is not overfitting.

```python
#plotting predicted v/s actual values
sns.distplot(preds,label='predicted values')
sns.distplot(y_1_test,label='actual values')
plt.title('predicted v/s actual values')
plt.legend()
plt.show()
```

# FINAL FITTING for AVERAGE PRICE PREDICTIONS

```python
#finally fiiting 100% data for deployment
KNN_best.fit(x_1,y_1)
KNN_best.score(x_1,y_1)
```

```python
#saving the model using joblib
import joblib 
joblib.dump(KNN_best,'avocado_avg_price.pkl')
```

# 2) Predicting Regions.

```python
#seperating the target variable
x_2=avoc_dat.drop('region',axis=1)
y_2=avoc_dat['region']
```

```python
#check input data
x_2.head()
```

```python
#standard scale input variables
x_2=pd.DataFrame(scale.fit_transform(x_2),columns=x_2.columns)
x_2.head()
```

```python
#train-test-split
x_2_train,x_2_test,y_2_train,y_2_test=train_test_split(x_2,y_2,test_size=0.3,random_state=4)
```

```python
#importing all the base classification algorithms
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
```

```python
#importing all the metrics for regression analysis
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,roc_curve
```

```python
#selecting the algorithm which works best 
model_list=[KNeighborsClassifier(),GaussianNB(),SVC(),DecisionTreeClassifier()]
for model in model_list:
    results=cross_val_score(model,x_2_train,y_2_train,cv=10,scoring='accuracy')
    print(model)
    print('mean score : ',results.mean())
    print('\n')
```

SVC and KNeighborsClassifier are giving better generalised scores.

# 1)Tuning SVC

```python
#tuning hyperparameters of svc
svc=SVC()
para = {'C' : [0.001,0.01,0.1,1],'kernel':['linear','poly','rbf']}
GCV_svc=GridSearchCV(estimator=svc,param_grid=para,cv=10)
GCV_svc.fit(x_2_train,y_2_train)
```

```python
#getting best parameters for svc
print('best parameters :',GCV_svc.best_params_)
```

```python
#training and testing with best parameters of svc
svc_best=SVC(C= 1, kernel='rbf')
svc_best.fit(x_2_train,y_2_train)
prediction=svc_best.predict(x_2_test)
print('accuracy_score :',accuracy_score(prediction,y_2_test))
print('confusion matrix :','\n',confusion_matrix(y_2_test,prediction))
print('classification report :','\n',classification_report(y_2_test,prediction))
```

```python
#generalised accuracy score using best parameters of svc
cv_scores=cross_val_score(svc_best,x_2_train,y_2_train,cv=10,scoring='accuracy')
print('mean score :',cv_scores.mean())
```

# 2) Tuning KNeighborsClassifier

```python
#tuning hyperparameters of KNeighborsClassifier
hyper_params={'n_neighbors':[3,5,7,9,10],'weights':['uniform','distance'],'metric':['euclidean','manhattan','minkowski']}
KNC=KNeighborsClassifier()
GCV_KNC=GridSearchCV(estimator=KNC,param_grid=hyper_params,cv=10)
GCV_KNC.fit(x_2_train,y_2_train)
```

```python
#getting best parameters for KNeighborsClassifier
print('best parameters :',GCV_KNC.best_params_)
```

```python
#training and testing with best parameters of KNeighborsClassifier
KNC_best=KNeighborsClassifier(n_neighbors=3,metric='manhattan',weights='distance')
KNC_best.fit(x_2_train,y_2_train)
predictions=KNC_best.predict(x_2_test)
print('accuracy_score :',accuracy_score(predictions,y_2_test))
print('confusion matrix :','\n',confusion_matrix(y_2_test,predictions))
print('classification report :','\n',classification_report(y_2_test,predictions))
```

```python
#generalised accuracy score using best parameters of KNeighborsClassifier
cv_score=cross_val_score(KNC_best,x_2_train,y_2_train,cv=10,scoring='accuracy')
print('mean score :',cv_score.mean())
```

We can now see that the f-1 scores of KNeighborsClassifier are better than f-1 scores of SVC.

Also, the generalised acuuracy score of KNeighborsClassifier is almost 8% better than SVC.

We can now conclude that KNeighborsClassifier works best!

```python
#plotting roc_curve
pred_prob=KNC_best.predict_proba(x_2_test)[:,1]
fpr,tpr,threshholds=roc_curve(y_2_test,pred_prob,pos_label=True)
plt.plot(fpr,tpr)
plt.xlabel('false positive rate')
plt.ylabel('true positive rate')
plt.title('roc_curve')
plt.show()
```

Looking at the curve , we can say that the model is performing excellently !

```python
#finally fitting 100% data for deployment
KNC_best.fit(x_2,y_2)
KNC_best.score(x_2,y_2)
```
