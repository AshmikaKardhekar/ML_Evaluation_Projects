```python
#check skewness of data
hr_dat.skew()
```

The skewness of all the continuous variables need to be improved.Rest all are discrete variables,we won't improve there skewness. Also all continuous variables are positively skewed.

Let us remove skewness where columns have skewness corfficient of more than 0.5

```python
#remove skewness of columns with skewness coeff more than 0.5 using log
for i in hr_dat.columns:
    if hr_dat[i].dtype=='int64':
        if hr_dat[i].skew() > 0.5:
            hr_dat[i]=np.log1p(hr_dat[i])
        if hr_dat[i].skew() < -0.5:
            hr_dat[i]=np.square(hr_dat[i])
```

```python
#check the improved skewness 
hr_dat.skew()
```

The skewness of all the continuous variables is improved.

```python
#encode target varible (attrition)
#import label encoder from sklearn
import sklearn
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
hr_dat['Attrition']=le.fit_transform(hr_dat['Attrition'])
```

```python
#plotting correlation heatmap
plt.figure(figsize=(20,10))
sns.heatmap(hr_dat.corr(),annot=True)
plt.show()
```

To avoid multi-colinearity, let us drop one feature from all the feature pairs which are highly correlated(more than 0.7) with each other :

We have high correlation between features joblevel and monthly income,but the magnitude of correlation of joblevel with the target variable(attrition) is less than that of monthly income. So let's drop the feature job level.

In similar way we can remove the features 'performance rating','years in current role','years with current manager'.

```python
#dropping the above mentioned features
hr_dat.drop(['JobLevel','PerformanceRating','YearsInCurrentRole','YearsWithCurrManager'],axis=1,inplace=True)
```

```python
#seperating target varibale
x=hr_dat.drop('Attrition',axis=1)
y=hr_dat['Attrition']
```

```python
#check object features in x
for o in x.columns:
    if x[o].dtype=='object':
        print(o)
```

```python
#encode all the categorical varibles in x
x=pd.get_dummies(x)
x.head()
```

```python
#visualising features
x.plot(kind='box',subplots=True,layout=(7,7),figsize=(20,20))
plt.show()
```

We see few columns with outliers.

But they are not significant , so we will ignore them.

```python
#scaling the input variables
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
x=pd.DataFrame(scale.fit_transform(x),columns=x.columns)
x.head()
```

# TRAINING AND TESTING THE MODEL

```python
#train-test-split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=4)
```

```python
#importing all the base classification algorithms
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
```

```python
#importing all the metrics for regression analysis
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,roc_curve,roc_auc_score
```

```python
#selecting the base algorithm which works best 
from sklearn.model_selection import cross_val_score
model_list=[LogisticRegression(),KNeighborsClassifier(),GaussianNB(),SVC(),DecisionTreeClassifier()]
for model in model_list:
    results=cross_val_score(model,x_train,y_train,cv=10,scoring='accuracy')
    print(model)
    print('mean score : ',results.mean())
    print('\n')
```

LogisticRegression and SVC are giving better generalised scores.

# 1) SVC

```python
#tuning hyperparameters of svc
from sklearn.model_selection import GridSearchCV
svc=SVC()
params = {
 'C' : [0.001,0.01,0.1,1,10,100],
 'kernel':['linear','poly','rbf']
 }
GCV_svc=GridSearchCV(estimator=svc,param_grid=params,cv=10)
GCV_svc.fit(x_train,y_train)
```

```python
#getting best parameters for svc
print('best parameters :',GCV_svc.best_params_)
```

```python
#training and testing with best parameters of svc
svc_best=SVC(C= 0.1, kernel= 'linear',probability=True)
svc_best.fit(x_train,y_train)
prediction=svc_best.predict(x_test)
print('accuracy_score :',accuracy_score(y_test,prediction))
print('confusion matrix :','\n',confusion_matrix(y_test,prediction))
print('classification report :','\n',classification_report(y_test,prediction))
print('roc_auc_score : ',roc_auc_score(y_test,prediction))
```

```python
#generalised accuracy score using best parameters of svc
cv_scores=cross_val_score(svc_best,x_train,y_train,cv=10,scoring='accuracy')
print('mean accuracy score :',cv_scores.mean())
```

```python
#generalised roc_auc score using best parameters of AdaBoost with LogisticRegressor as base estimator
cv_scores=cross_val_score(svc_best,x_train,y_train,cv=10,scoring='roc_auc')
print('mean roc_auc score :',cv_scores.mean())
```

```python
#plotting roc_curve of LogisticRegression .
pred_pro=svc_best.predict_proba(x_test)[:,1]
fpr,tpr,threshholds=roc_curve(y_test,pred_pro,pos_label=True)
plt.plot(fpr,tpr)
plt.xlabel('false positive rate')
plt.ylabel('true positive rate')
plt.title('roc_curve')
plt.show()
```

# 2) LogisticRegression 

```python
#tuning hyperparameters of LogisticRegressor
lr=LogisticRegression()
param = {
 'penalty': ['l1','l2'],
 'C' : [0.001,0.01,0.1,1,10,100,1000],
 'solver':['newton-cg','lbfgs','liblinear','sag','saga']
 }
GCV_lr=GridSearchCV(estimator=lr,param_grid=param,cv=10)
GCV_lr.fit(x_train,y_train)
```

```python
#getting best parameters for LogisticRegressor
print('best parameters :',GCV_lr.best_params_)
```

```python
#training and testing with best parameters of LogisticRegressor
lr_best=LogisticRegression(C= 0.1, penalty= 'l2', solver= 'newton-cg')
lr_best.fit(x_train,y_train)
predictions=lr_best.predict(x_test)
print('accuracy_score :',accuracy_score(y_test,predictions))
print('confusion matrix :','\n',confusion_matrix(y_test,predictions))
print('classification report :','\n',classification_report(y_test,predictions))
print('roc_auc_score : ',roc_auc_score(y_test,predictions))
```

```python
#generalised accuracy score using best parameters of LogisticRegressor
cv_scores=cross_val_score(lr_best,x_train,y_train,cv=10,scoring='accuracy')
print('mean accuracy score :',cv_scores.mean())
```

```python
#generalised roc_auc score using best parameters of AdaBoost with LogisticRegressor as base estimator
cv_scores=cross_val_score(lr_best,x_train,y_train,cv=10,scoring='roc_auc')
print('mean roc_auc score :',cv_scores.mean())
```

```python
#plotting roc_curve of LogisticRegression .
pred_prob=lr_best.predict_proba(x_test)[:,1]
fpr,tpr,threshholds=roc_curve(y_test,pred_prob,pos_label=True)
plt.plot(fpr,tpr)
plt.xlabel('false positive rate')
plt.ylabel('true positive rate')
plt.title('roc_curve')
plt.show()
```

# Final Selection betweeen LogisticRegression and SVC

After analysing both the algorithms we came to a conclusion that both these algorithm are working almost equally. So, we can go ahead with any of these two algorithms for final training.

Selecting LogisticRegression as Final base model.(As mean roc_auc score is slightly better than that of SVC).

let us try to improve the model performance by boosting algorithm.

```python
#boosting the algorithm using AdaBoost and tuning it.
from sklearn.ensemble import AdaBoostClassifier
ada_lr=AdaBoostClassifier(base_estimator=lr_best)
para = {
 'n_estimators': [50, 100,150,200,250,300],
 'algorithm':['SAMME','SAMME.R']
 }
GCV_ada_lr=GridSearchCV(estimator=ada_lr,param_grid=para,cv=10)
GCV_ada_lr.fit(x_train,y_train)
```

```python
#getting best parameters of AdaBoost 
print('best parameters :',GCV_ada_lr.best_params_)
```

```python
#training and testing with best parameters of AdaBoostClassifier
ada_lr_best=AdaBoostClassifier(base_estimator=lr_best,algorithm='SAMME.R',n_estimators=50)
ada_lr_best.fit(x_train,y_train)
preds=ada_lr_best.predict(x_test)
print('accuracy_score :',accuracy_score(y_test,preds))
print('confusion matrix :','\n',confusion_matrix(y_test,preds))
print('classification report :','\n',classification_report(y_test,preds))
print('roc_auc_score :',roc_auc_score(y_test,preds))
```

Looking at the curve , we can say that the model is performing good!

```python
#generalised accuracy score using best parameters of AdaBoost with LogisticRegressor as base estimator
cv_scores=cross_val_score(ada_lr_best,x_train,y_train,cv=10,scoring='accuracy')
print('mean accuracy score :',cv_scores.mean())
```

```python
#generalised roc_auc score using best parameters of AdaBoost with LogisticRegressor as base estimator
cv_scores=cross_val_score(ada_lr_best,x_train,y_train,cv=10,scoring='roc_auc')
print('mean roc_auc score :',cv_scores.mean())
```

After using Adaboost the mean accuracy and roc_auc scores are decresed by small amount.It is better we only use LogisticRegression for final fitting.

```python
#finally fitting 100% data for deployment
lr_best.fit(x,y)
lr_best.score(x,y)
```
