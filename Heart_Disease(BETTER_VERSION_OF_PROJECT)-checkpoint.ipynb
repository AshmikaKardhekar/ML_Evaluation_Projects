# Import needed modules and get the required dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# get the dataset
heart_dat = pd.read_csv('Downloads/processed.cleveland.data',names=['age','sex','cp','trestbps','chol','fbs',
                                                                    'restecg','thalach','exang','oldpeak','slope','ca','thal',
                                                                    'num'])
heart_dat.head()
```

FEATURE DESCRIPTION :

1.  age: age in years
2.  sex: sex (1 = male; 0 = female)
3.  cp: chest pain type
        -- Value 1: typical angina
        -- Value 2: atypical angina
        -- Value 3: non-anginal pain
        -- Value 4: asymptomatic
4.  trestbps: resting blood pressure (in mm Hg on admission to the hospital)
5.  chol: serum cholestoral in mg/dl
6.  fbs: (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)
7.  restecg: resting electrocardiographic results
        -- Value 0: normal
        -- Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)
        -- Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria
8.  thalach: maximum heart rate achieved
9.  exang: exercise induced angina (1 = yes; 0 = no)
10. oldpeak: ST depression induced by exercise relative to rest
11. slope: the slope of the peak exercise ST segment
        -- Value 1: upsloping
        -- Value 2: flat
        -- Value 3: downsloping
12. ca: number of major vessels (0-3) colored by flourosopy
13. thal: 3 = normal; 6 = fixed defect; 7 = reversable defect

TARGET DESCRIPTION :


14. num: presence of disease (values 1,2,3,4) from absence (value 0).

```python
#seperating the target variable
x=heart_dat.drop('num',axis=1)
y=heart_dat['num']
```

```python
#plotting correlation heatmap
plt.figure(figsize=(12,6))
sns.heatmap(heart_dat.corr(),annot=True)
plt.show()
```

fbs and chol are the lowest correlated with the target variable, ca, thal, oldpeak, are in better correlation with target variable than other features.

Also, we can see that there are many variables which are not ordinal and hence must be one-hot-encoded, otherwise the categories in most of the features with numeric label such as 0,1,2,3,... will be given weights according to their actual numerical weights, by any algorithm.

example : feature 'thal' (3 = normal; 6 = fixed defect; 7 = reversable defect) , so 3(normal) will be given less weight than 6(fixed defect) and 6(fixed defect) will get less weight than 7(reversable defect), by any algorithm, but 3(normal),6(fixed defect),7(reversable defect) are independent of each other and hence must be provided equal weights by the model. Hence, to avoid this we one-hot-encode these features using pandas.

```python
#encode all the categorical varibles in x
cat_var = ['sex','cp','fbs','restecg','exang','slope','ca','thal']
x=pd.get_dummies(x,columns=cat_var)
x.head()
```

```python
#visualising columns in input data
x.hist(grid=True,bins=20,layout=(4,7),figsize=(20,10))
plt.show()
```

most of the columns show abnormalities, let's check skewness of all columns.

```python
#check skewness of all columns in x.
x.skew()
```

```python
#remove skewness of columns with skewness coeff more than 0.5 using log and square
for i in x.columns:
    if x[i].skew() > 0.5:
        x[i]=np.log1p(x[i])
    if x[i].skew() < -0.5:
        x[i]=np.square(x[i])
x.skew() #check improved skewness
```

The skewness of all continuous variables with skewness coeff more than 0.5 magnitude, is improved now.

```python
#visualising columns after improving skewness
x.plot(kind='box',subplots=True,layout=(4,7),figsize=(20,10))
plt.show()
```

Outliers are not much, so we will ignore them.

```python
#standard scale input variables
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
x=pd.DataFrame(scale.fit_transform(x),columns=x.columns)
x.head() #check data
```

```python
#importing all the base classification algorithms
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
```

```python
#importing all the metrics for regression analysis
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,roc_curve
```

```python
#import train-test-split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=None)
#import cross val score 
from sklearn.model_selection import cross_val_score
model_list=[KNeighborsClassifier(),GaussianNB(),SVC(),DecisionTreeClassifier()]
for model in model_list:
    results=cross_val_score(model,x_train,y_train,cv=10,scoring='accuracy')
    print(model)
    print('mean score : ',results.mean())
    print('\n')
```

SVC and KNeighbours are working well.

```python
#checking the accuracy score of SVC at best random_state
svc=SVC()
max_score_svc=0
best_ran_state_svc=0
for n in range(0,101):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=n)
    svc.fit(x_train,y_train)
    pred=svc.predict(x_test)
    score=accuracy_score(y_test,pred)
    if max_score_svc < score:
        max_score_svc = score
        best_ran_state_svc = n
print('max accuracy score of svc : ',max_score_svc)
print('best random state : ',best_ran_state_svc)
```

```python
#checking the mean accuracy score of KNeighbors at best random_state
KNC=KNeighborsClassifier()
max_score_knc=0
best_ran_state_knc=0
for n in range(0,101):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=n)
    KNC.fit(x_train,y_train)
    pred=KNC.predict(x_test)
    score=accuracy_score(y_test,pred)
    if max_score_knc < score:
        max_score_knc = score
        best_ran_state_knc = n
print('max accuracy score of KNC: ',max_score_knc)
print('best random state : ',best_ran_state_knc)
```

Both KNN and SVC are working well with KNN working better than SVC.

Let's check the tuned performances of both the algorithms.

# KNN

```python
#train test split with best random state obtained for KNN
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=98)
#importing GridSearchCV 
from sklearn.model_selection import GridSearchCV
#tuning hyperparameters of KNeighborsClassifier
hyper_params={'n_neighbors':range(3,15),
              'weights':['uniform','distance'],
              'metric':['euclidean','manhattan','minkowski']}
GCV_KNC=GridSearchCV(estimator=KNC,param_grid=hyper_params,cv=10)
GCV_KNC.fit(x_train,y_train)
```

```python
#getting best parameters for KNeighborsClassifier
print('best parameters :',GCV_KNC.best_params_)
```

```python
#training and testing with best parameters of KNeighborsClassifier
KNC_best=KNeighborsClassifier(n_neighbors=11,metric='euclidean',weights='uniform')
KNC_best.fit(x_train,y_train)
predictions=KNC_best.predict(x_test)
print('accuracy_score :',accuracy_score(y_test,predictions))
print('confusion matrix : ','\n',confusion_matrix(y_test,predictions))
print('classification report : ','\n',classification_report(y_test,predictions))
```

```python
#plotting roc_curve 
pred_prob=KNC_best.predict_proba(x_test)[:,1]
fpr,tpr,threshholds=roc_curve(y_test,pred_prob,pos_label=True)
plt.plot(fpr,tpr)
plt.xlabel('false positive rate')
plt.ylabel('true positive rate')
plt.title('roc_curve')
plt.show()
```

```python
#generalised score of KNN
results_knc=cross_val_score(KNC_best,x,y,cv=10,scoring='accuracy')
results_knc.mean()
```
